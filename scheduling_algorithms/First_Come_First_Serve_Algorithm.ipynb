```python
class node:
    def __init__(self):
        self.process = None
        self.next = None

class process:
    
    def __init__(self,burst_time,arrival_time=None,priority=None):
        self.burst_time = burst_time
        if arrival_time == None:
            arrival_time = time.time()
        self.arrival_time = arrival_time
        self.priority = priority
        self.last_ran = None
        
class que:
    
    def __init__(self):
        self.first = None
        self.time = time.time()
        self.size = 0
    
    # x is a process
    def FCFS_insert(self,x):
        if self.first == None :
            self.first = node()
            self.first.process = x
            x.last_ran = x.arrival_time
            self.size = 1
        else:
            
            n = self.first
            l = n
            
            while n != None:
                l = n
                n = n.next
                
            l.next = node()
            l.next.process = x
            x.last_ran = x.arrival_time
            self.size +=1
    
    # update time of que
    def time_update(self,time):
        self.time = time
        
    # update function for First Come First  method
    def FCFS_update(q,now):
        #print('>')
        if q.first == None:
            q.time = now
            return

        fp = q.first
        if fp.process.burst_time + fp.process.last_ran > now:
            fp.process.burst_time -= now - fp.process.last_ran
            fp.process.last_ran = now
            q.time = now
        else:
            q.first = fp.next
            if q.first != None : 
                q.first.process.last_ran = fp.process.last_ran +fp.process.burst_time
            q.time = fp.process.last_ran +fp.process.burst_time
            FCFS_update(q,now)

        return
    
    def update(self,now,which='FCFS'):
        
        if which == 'FCFS':
            FCFS_update(self,now)
        
    # utility function to see the que
    def show(self):
        f = q.first
        print('at time :', self.time)
        while f != None:
            print('burst:',f.process.burst_time,"| arrival:",f.process.arrival_time,"| last :",f.process.last_ran )
            f = f.next
```
